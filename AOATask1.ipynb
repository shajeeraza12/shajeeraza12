**The goal of this work is to experimentally analyze the time complexity of different algorithms by measuring execution times for various operations on vectors and matrices. The results will help compare the theoretical and practical performance of these algorithms.**

**The problem involves generating random vectors and matrices of varying sizes and performing different operations such as sum, product, sorting, and matrix multiplication. The execution times of these operations will be measured and analyzed to study their time complexities and compare theoretical expectations with practical results.**



```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

# 1) ùëì(ùíó) = ùëêùëúùëõùë†ùë° (constant function);
# 2) ùëì(ùíó) = ‚àë ùë£ (the sum of elements);
# 3) ùëì(ùíó) = ‚àè ùë£ (the product of elements);

The summation function calculates the total sum of elements in a vector, and has a time complexity of O(n). The multiplication function computes the product of vector elements, also with a time complexity of O(n). The constant function, which returns a fixed value regardless of the input, has a time complexity of O(1), as its execution time is constant.

```python
# Function to generate an n-dimensional random vector
def generate_vector(n):
    return np.random.rand(n)

# Constant function that returns a constant value
def constant_function(vector):
    return 42

# Sum function that returns the sum of elements in the vector
def sum_function(vector):
    return np.sum(vector)

# Product function that returns the product of elements in the vector
def product_function(vector):
    return np.prod(vector)
```

```python
# Initialize storage for execution times
execution_times_constant = []
execution_times_sum = []
execution_times_product = []
```

```python
# Measure execution times for each function for each n in a single loop
for n in range(1, 2001):
    vector = generate_vector(n)

    # Measure time for all three functions in one loop
    start_time = time.perf_counter()
    for _ in range(5):
        _ = constant_function(vector)
    end_time = time.perf_counter()
    average_time_constant = (end_time - start_time) / 5
    execution_times_constant.append(average_time_constant)

    start_time = time.perf_counter()
    for _ in range(5):
        _ = sum_function(vector)
    end_time = time.perf_counter()
    average_time_sum = (end_time - start_time) / 5
    execution_times_sum.append(average_time_sum)

    start_time = time.perf_counter()
    for _ in range(5):
        _ = product_function(vector)
    end_time = time.perf_counter()
    average_time_product = (end_time - start_time) / 5
    execution_times_product.append(average_time_product)

# Plot the execution times for all three functions
plt.figure(figsize=(10, 6))
plt.plot(range(1, 2001), execution_times_constant, label="Constant Function", color="blue")
plt.plot(range(1, 2001), execution_times_sum, label="Sum Function", color="green")
plt.plot(range(1, 2001), execution_times_product, label="Product Function", color="red")

# Set labels and title
plt.xlabel("Vector Dimension (n)")
plt.ylabel("Average Execution Time (seconds)")
plt.title("Execution Time Comparison of Functions: Constant, Sum, Product")
plt.legend()
plt.grid(True)


plt.show()
```

The experimental results for the constant function align with the theoretical expectation of ùëÇ(1), showing a constant execution time. Both the summation and multiplication functions follow the expected O(n) time complexity, with execution time increasing linearly with the vector size. While small deviations and spikes are observed, especially for multiplication, these could be caused by system load, memory caching, or the overhead of running the functions multiple times. However, the overall trend still confirms the theoretical complexities of O(1) for constant and O(n) for summation and multiplication.

# Sorting Algorithms
# 1) Bubble Sort
# 2) Quick Sort
# 3) Time Sort

The sorting algorithms used in this task are Bubble Sort, Quick Sort, and Timsort. Bubble Sort repeatedly swaps adjacent elements to arrange them in order with O(n^2) complexity, while Quick Sort uses a divide-and-conquer approach with O(nlogn) average complexity, and Timsort combines merge and insertion sorts, achieving O(nlogn) in most cases.

```python
# Bubble Sort implementation
def bubble_sort(vector):
    n = len(vector)
    for i in range(n):
        for j in range(0, n-i-1):
            if vector[j] > vector[j+1]:
                vector[j], vector[j+1] = vector[j+1], vector[j]

# Quick Sort implementation
def quick_sort(vector):
    if len(vector) <= 1:
        return vector
    pivot = vector[len(vector) // 2]
    left = [x for x in vector if x < pivot]
    middle = [x for x in vector if x == pivot]
    right = [x for x in vector if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Timsort uses Python's built-in sorted() which is Timsort
def timsort(vector):
    return sorted(vector)

# Initialize storage for execution times
execution_times_bubble_sort = []
execution_times_quick_sort = []
execution_times_timsort = []
```

```python
# Measure execution times for each sorting algorithm for each n in a single loop
for n in range(1, 2001):
    vector = generate_vector(n)

    # Measure time for bubble sort
    vector_copy = vector.copy()
    start_time = time.perf_counter()
    for _ in range(5):
        bubble_sort(vector_copy)
    end_time = time.perf_counter()
    average_time_bubble_sort = (end_time - start_time) / 5
    execution_times_bubble_sort.append(average_time_bubble_sort)

    # Measure time for quick sort
    vector_copy = vector.copy()
    start_time = time.perf_counter()
    for _ in range(5):
        quick_sort(vector_copy)
    end_time = time.perf_counter()
    average_time_quick_sort = (end_time - start_time) / 5
    execution_times_quick_sort.append(average_time_quick_sort)

    # Measure time for timsort (built-in Python sort)
    vector_copy = vector.copy()
    start_time = time.perf_counter()
    for _ in range(5):
        timsort(vector_copy)
    end_time = time.perf_counter()
    average_time_timsort = (end_time - start_time) / 5
    execution_times_timsort.append(average_time_timsort)

# Plot the execution times for all sorting algorithms
plt.figure(figsize=(12, 8))
plt.plot(range(1, 2001), execution_times_bubble_sort, label="Bubble Sort", color="orange")
plt.plot(range(1, 2001), execution_times_quick_sort, label="Quick Sort", color="purple")
plt.plot(range(1, 2001), execution_times_timsort, label="Timsort", color="brown")

# Set labels and title
plt.xlabel("Vector Dimension (n)")
plt.ylabel("Average Execution Time (seconds)")
plt.title("Execution Time Comparison of Sorting Algorithms (Bubble Sort, Quick Sort, Timsort)")
plt.legend()
plt.grid(True)


plt.show()
```

The experimental results show that Bubble Sort exhibits a significant increase in execution time, especially for larger vectors, consistent with its O(n^2) time complexity. In contrast, Quick Sort and Timsort demonstrate more efficient performance with execution times that increase much more gradually, in line with their O(nlogn) complexity.

# 4) supposing that the elements of ùíó are the coefficients of a polynomial ùëÉ of  degree ùëõ ‚àí1,  calculate the value ùëÉ(1.5) by a direct calculation of ùëÉ(ùë•) =‚àë ùë£ùë• (i.e. evaluating each term one by one) and by Horner‚Äôs method by representing the polynomial as P(ùë•) = ùë£ +ùë•ùë£ +ùë•(ùë£ +‚ãØ) ;

Direct calculation of a polynomial evaluates each term individually, with a time complexity of O(n) since each element of the vector is multiplied by a power of x. Horner's method, a more efficient approach, rewrites the polynomial to minimize redundant calculations, also with O(n) complexity but generally faster in practice due to fewer operations.

```python
# Direct polynomial evaluation (sum of terms v_k * x^(k-1))
def direct_polynomial_evaluation(v, x=1.5):
    result = 0
    n = len(v)
    try:
        for k in range(1, n + 1):
            term = v[k-1] * (x ** (k - 1))
            result += term
    except OverflowError:
        print(f"Overflow occurred for n = {n}, x = {x}")
        result = np.inf
    return result

# Horner's method for polynomial evaluation
def horners_method(v, x=1.5):
    result = v[-1]
    for k in range(len(v) - 2, -1, -1):
        result = v[k] + x * result
    return result

# Initialize lists to store the execution times for each method
execution_times_direct = []
execution_times_horner = []
```

```python
# Measure execution times for both methods for each n
for n in range(1, 2001):
    vector = generate_vector(n)

    # Measure time for direct polynomial evaluation
    start_time = time.perf_counter()
    for _ in range(5):
        _ = direct_polynomial_evaluation(vector)
    end_time = time.perf_counter()
    average_time_direct = (end_time - start_time) / 5
    execution_times_direct.append(average_time_direct)

    # Measure time for Horner's method
    start_time = time.perf_counter()
    for _ in range(5):
        _ = horners_method(vector)
    end_time = time.perf_counter()
    average_time_horner = (end_time - start_time) / 5
    execution_times_horner.append(average_time_horner)

# Plot the execution times for both methods
plt.figure(figsize=(10, 6))
plt.plot(range(1, 2001), execution_times_direct, label="Direct Method", color="blue")
plt.plot(range(1, 2001), execution_times_horner, label="Horner's Method", color="red")
plt.xlabel("Vector Dimension (n)")
plt.ylabel("Average Execution Time (seconds)")
plt.title("Execution Time Comparison: Direct Method vs Horner's Method")
plt.legend()
plt.grid(True)
plt.show()
```

The experimental results show that both the direct calculation and Horner's method exhibit linear growth in execution time, consistent with their O(n) complexity. However, the direct calculation occasionally leads to overflow errors due to large intermediate values when computing powers of x, whereas Horner‚Äôs method avoids such issues by reducing the number of multiplications and powers. Additionally, as predicted by theory, Horner's method takes less time in practice, as it minimizes redundant calculations, which was confirmed by the experimental results.
